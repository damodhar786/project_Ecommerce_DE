```sql
CREATE OR REPLACE VIEW ecommerce.gold.fact_transactions_denorm  AS (
SELECT gfo.*,
gdd.year,
gdd.month_name,
gdd.day_name,
gdd.is_weekend,
gdd.quarter,
gdd.week,
gdp.sku,
gdp.category_code,
gdp.category_name,
gdp.brand_code,
gdp.brand_name,
gdp.color,
gdp.size,
gdp.rating_count,
extract(Hour FROM transaction_ts) as hour_of_day
FROM ecommerce.gold.gld_fact_order_items gfo 
JOIN ecommerce.gold.gld_dim_date gdd
ON gfo.date_id = gdd.date_id
JOIN ecommerce.gold.gld_dim_products gdp
ON gfo.product_id = gdp.product_id
);
```
